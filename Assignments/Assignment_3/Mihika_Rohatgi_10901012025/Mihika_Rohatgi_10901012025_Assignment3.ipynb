```python
#Q1. Create a function to print first 10 natural numbers
def printno(n):
  for i in range(1,n+1):
    print(i)
printno(10)
```

```python
#Q2. Create a function to calculate sum of first N natural numbers
def sumN(N):
  sum=0
  for i in range(1,N+1):
    sum+=i
  return sum
sumN(10)
```

```python
#3. Create a function to reverse a number
def revNum(n):
  s=str(n)
  n=s[::-1]
  return int(n)
revNum(12345)
```

```python
#4. Create a function to count digits in a number
def count(n):
  s=str(n)
  return len(s)
count(1234567)
```

```python
#5. Create a function to check palindrome number
def pal(n):
  s=str(n)
  l=len(s)
  count=0
  for i in range(0,int(l/2)):
    if s[i]==s[l-i-1]:
      count+=1
    else:
      print("Number is not a palindrome")
      break
  if count==int(l/2):
    print("Number is a palindrome")
pal(12341)
```

```python
#6. Create a function to generate fibonacci series
def fib(n):
  a=0
  b=1
  print(a)
  print(b)
  for i in range(1,n-1):
    c=a+b
    a=b
    b=c
    print(c)
fib(10)
```

```python
#7. Calculator using functions
def add(a,b):
  print(a+b)
def sub(a,b):
  print(a-b)
def prod(a,b):
  print(a*b)
def div(a,b):
  print(a/b)
#menu-driven program
a=int(input("Enter first number:"))
b=int(input("Enter second number:"))
print("1. Addition")
print("2. Subtraction")
print("3. Product")
print("4. Divide")
n=int(input("Enter your choice:"))
if n==1:
  add(a,b)
elif n==2:
  sub(a,b)
elif n==3:
  prod(a,b)
elif n==4:
  div(a,b)
else:
  print("Enter correct choice.")
```

```python
#8. Create a text file and store student details.
n=int(input("Enter number of students:"))
for i in range(0,n):
  name=input("Enter student name:")
  enrollno=int(input("Enter enrollment number:"))
  branch=input("Enter branch name:")
  college=input("Enter college name:")
  with open("students.txt","a") as file:
    file.write(f"{name},{enrollno},{branch},{college}\n")
  print("Student details stored successfully!")
```

```python
#Q11. Create a Student class with name and marks.
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
    def display_details(self):
        print(f"Student Name: {self.name}")
        print(f"Marks: {self.marks}")
student1 = Student("Mihika", 95)
student1.display_details()
```
